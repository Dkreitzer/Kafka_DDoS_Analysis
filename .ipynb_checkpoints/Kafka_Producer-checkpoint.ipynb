# Kafka Producer

### This notebook:
 - Reads and parses a local text file,
 - Pushes (KafkaProducer) individual lines of the text file to Kafka

```python
from json import dumps
from kafka import KafkaProducer
import os
import re
```

### Load and open local text file

```python
# Raw Data - unzipped gz file
rawData = 'data/apache-access-log.txt'
textFile = open(rawData, "r")
```

### Parse and split the text file

```python
Lines = []

for line in textFile:
    line.rstrip()
    line = re.sub(r'[-""]', '', line)
    entry = line.split()
    Lines.append(entry)
    
print(f'Total items in Lines list: {len(Lines)}')

print('')
print(f'Example of item 0 in Lines list: {Lines[0]}')
```

### Create Kafka Producer

```python
producer = KafkaProducer(bootstrap_servers=['localhost:9092'],value_serializer=lambda x: dumps(x).encode('utf-8'))
```

### Push (KafkaProducer) objects to Kafka

#### IMPORTANT:
 - Verify and update if needed line 3 with the Kafka Topic name you are sending data to

```python
odometer = 0
for item in Lines:
    producer.send('ipTest9', value=item)
    print('one line sent')
    odometer = odometer +1
print(f"number of files sent  {odometer}")    
```
